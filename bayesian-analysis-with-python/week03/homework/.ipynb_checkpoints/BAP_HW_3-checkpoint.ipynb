### Exercise 1.2: Dircihlet distribution first moment (5 pts)

* Write the Dirihlet distribution in expential family form, where pdf of the Dirichlet is given by:


$$\lrg{ 
p(x | \alpha) = \frac{\Gamma(\sumK \alpha_k)}{\prod_{k=1}^K\Gamma(\alpha_k)}\prod_{k=1}^K x_k^{\alpha_k-1}, \where x_k \in (0,1)\ \text{for}\ k = \overline{1..K},  \and \sumK{x_k} = 1.
}$$
<br><br>
Let's define $$\lrg{\eta = \alpha - 1 = (\alpha_1-1,...,\alpha_K-1).}$$
One immediately sees that one can define $g$ as 
$$
\lrg{
g(\eta) = g(\alpha - 1) = \frac{\Gamma(\sum \alpha_k)}{\Pi (\Gamma(\alpha_k))}
}$$
and
$$\lrg{
h(x) = 1 \quad \forall x\in \mathbb R^K.}
$$

$$\lrg{ 
p(x | \eta) = g(\eta) \exp(\ln(\Pi x_k^{\alpha_k - 1})) = g(\eta) \exp(\sum (\alpha_k - 1) \underbrace{\ln x_k}_{=u_k(x_k)}) = g(\eta) \exp(\eta^T u(x))
}$$

So $$\lrg{
u(x) = (u_1(x_1),...,u_K(x_K)) = (\ln x_1, ..., \ln x_k).}
$$

* Using the exponential family property show that the first moment of the Dirichlet distribution is :


$$\lrg{\E[\ln x_k] = \Psi(\alpha_k) - \Psi\Big(\sum_{k=1}^K \alpha_k \Big)
}$$


**Property:** $\lrg{
\E[u(x)] = -\triangledown_{\eta} \ln g(\eta)  = \triangledown_{\eta} A(\eta)
}$

**Note:** Digamma function $\lrg{ \Psi(x) =  \frac{d}{dx} \ln \Gamma(x) = \frac{\Gamma(x)^{'}}{\Gamma(x)}}$ 


$$
\lrg{
     \E[(u(x)] = \E[\ln x_k] =  - \nabla_{x_k} \ln(\Gamma \sum x_k ) + \nabla_{x_k} \ln(\Pi \Gamma x_k) = - \Psi(\sum x_k) + \nabla_{x_k} \ln \Gamma x_k = - \Psi(\sum x_k) + \Psi(x_k).
}
$$

### Exercise 1.4: Multinomial-Dirichlet conjugacy (5 pts).

* Evaluate the posterior of $\theta$ given the model distribution is multinomial with $\theta$ and prior is Dirichlet with $\alpha$. See the picture below.
<br>

<img src='mdir_conj.png' width=400>

Model distribution

$$\lrg{
p(x_n | \theta) \sim \Mult(x_n | \theta) ,\quad \ln p(x_n | \theta) = \sumK x_{nk} \ln \theta_{k} + \const
}$$

Prior distribution

$$\lrg{
p(\theta | \alpha) = \Dir(\theta | \alpha), \quad \ln p(\theta | \alpha) = \ln\C(\alpha) + \sumK (\alpha_k - 1) \ln\theta_{k}, \where \\ \quad \quad \quad \quad  \quad \quad  \quad  \ln\C(\alpha) = \ln\Gamma\Big(\sumK \alpha_k\Big) - \sumK \ln\Gamma\Big(\alpha_{k}\Big)
}$$

* Evaluate the posterior distribution

$$\lrg{
\ln p(\theta | X, \alpha) = \ln p(X | \theta) + \ln p(\theta | \alpha) + \const 
}$$
<br><br>
$$\lrg{
\ln p(\theta | X, \alpha) = 
   \sum^K_{k=1}(\sum_{x_n \in X}x_{n,k} \ln \Theta_k) + \ln \mathcal C(\alpha) + \sum^K_{k=1}(\alpha_k - 1) \ln \Theta_k
}$$
Then, we take everything to the power of $e$, and we obtain
$$\lrg{\begin{align}
p(\theta | X, \alpha) &=
    \exp(\sum^K_{k=1}(\sum_{x_n \in X} x_{n,k} \ln \Theta_k)) \cdot \mathcal C(\alpha) \cdot \exp(\sum^K_{k=1}(\alpha_k - 1) \ln \Theta_k)\\
    &=
    \exp(\sum^K_{k=1}\left((\sum_{x_n \in X} x_{n,k}) + \alpha_k - 1\right) \ln \Theta_k)\cdot \mathcal C(\alpha)
\end{align}}
$$
We see this is in the exponential form of the dirichlet distribution with $$\lrg{\hat \alpha_k = \sum_{x_n \in X} x_{n,k} + \alpha_k}.$$
In other words,
$$
\lrg{
   p(\theta | X, \hat \alpha) = \lambda \cdot \mathcal C(\alpha) \prod_{k=1}^K x_k^{\sum_{x_n \in X}x_{n,k} + \alpha_k-1}
}
$$
for a factor $\lambda \in \mathbb R$ such that $\lambda C(\alpha) = C(\hat \alpha)$.
<br><br>
$$\boxed{\lrg{
p(\theta| X, \alpha) \sim \Dir(\theta | \hat\alpha),\where \hat\alpha = (\sum_{x_n \in X}x_{n,k} + \alpha_k)_{k=1,...,K}
}}$$

### Exercise 1.5: Dirichlet posterior (3 pts)

* Implement the posterior pdf as a function of input data X and hyperparameter $\alpha$.


* Your posterior function should return computed pdf value as well as the alpha_hat.


* You may use **dirichlet** function from _scipy.stats_ package for pdf computation.


```python
def posterior(x, X, alpha):    
    alpha_hat = alpha + np.sum(X, axis=0)
    pdfs = dirichlet.pdf(x,alpha_hat)
    return pdfs, alpha_hat
```

### Exercise 1.7: Evidence function (5 pts)

* Evaluate the negative evidence function $-\mathcal{L}(\alpha) = -\ln p(X|\alpha)$ as a function of the hyperparameter $\alpha$ and data $X$.


$$\lrg{
p(X | \alpha) = \int p(X, \theta|\alpha)\ d\theta = \int p(X | \theta) p (\theta | \alpha)\ d\theta  
}$$
It holds
$$
 \int p(X | \theta) p (\theta | \alpha)\ d\theta = \int \prod_{x_i \in X} \frac{\Gamma(n+1)}{\Pi^K_{j=1} \Gamma (x_{ij} + 1)} \prod_{j=1}^K \theta_j^{x_{ij}} \frac{\Gamma(\sum^K_{j=1} \alpha_j)}{\prod^K_{j=1} \Gamma(\alpha_j)}d \theta_1,...,\theta_K =
  \prod_{x_i \in X} \frac{\Gamma(n+1)}{\Pi^K_{j=1} \Gamma (x_{ij} + 1)} \frac{\Gamma(\sum^K_{j=1} \alpha_j)}{\prod^K_{j=1} \Gamma(\alpha_j)} \int \prod^K_{j=1}\theta_j^{\sum_{x_i \in X} x_{ij} + \alpha_j - 1}d \theta_1,...,\theta_K =
    \frac{\Gamma(n+1)}{\prod_{x_i \in X}\Pi^K_{j=1} \Gamma (x_{ij} + 1)} \frac{\Gamma(\sum^K_{j=1} \alpha_j)}{\prod^K_{j=1} \Gamma(\alpha_j)} \frac{\prod^K_{j=1}\Gamma(\sum_{x_i \in X} x_{ij} + \alpha_j)}{\Gamma(|\{ x_i \in X \}| N + \sum^K_{j=1} \alpha_j)}
$$
<br><br>
$$\lrg{
-\ln p(X | \alpha) =  -\ln(\frac{\Gamma(n+1)}{\prod_{x_i \in X}\Pi^K_{j=1} \Gamma (x_{ij} + 1)}) - \ln(\frac{\Gamma(\sum^K_{j=1} \alpha_j)}{\prod^K_{j=1}) \Gamma(\alpha_j)} -\ln\frac{\prod^K_{j=1}\Gamma(\sum_{x_i \in X} x_{ij})
\alpha_j)}{\Gamma(|\{ x_i \in X \}| N - \sum^K_{j=1} \alpha_j)}) }$$

* Implement the negative **evidence** function

```python
from scipy.special import loggamma
def neg_evidence(alpha,X):
    # YOUR CODE HERE
    neg_ev = loggamma(np.sum(alpha)) - np.sum(loggamma(alpha)) + loggamma(N+1)*25 - np.sum(loggamma(X+1)) + np.sum(loggamma(X+alpha)) - loggamma(N*25 + np.sum(alpha))
    
    return neg_ev 
```

### Exercise 1.8: Bayesian empirical learning (5 pts)

* Perform hyperparameter $\alpha$ minimization using the negative evidence function with _scipy.minimize_ function with the given boundaries.


* Collect and print the best (rounded to integer) alphas after minimization.

```python
from scipy.optimize import minimize

bnds = np.array([[0.01,np.inf]]*K)

best_alphas = []
for x in all_X:
    
    # YOUR CODE HERE
    res = minimize(fun=neg_evidence, x0=alpha_0, args=(all_X), bounds = bnds)
    
    best_alpha = res['x']
    best_alphas.append(best_alpha)
    
# YOUR CODE HERE
print(best_alpha)
print(best_alpha.astype(int))
```

### Exercise 2.2: Topic model toy data generation (10 pts)

* Implement function called **toy** which generates all of the LDA model parameters: $W$, $z_{true}$ and $\theta_{true}$ for the given (true) betas. Use $\alpha_0 = 0.1$ as a defult value.


* Input parameters: 
    
    *$M$-number of the generated document
    
    *$N$-number of the obseved words in each document
    
    
* Use **multinomial** function from _scipy.stats_ package with random seed 42.


**Note:** number of topics $K$ and vocabulary size $V$ can be retrieved from the true betas.

```python
from scipy.stats import multinomial, dirichlet

def toy(M, N, beta, alpha_0=.1):
    K,V = beta.shape # K=dim(Topics), V=dim(Vocabulary)
    Z = np.empty((M,N,K),dtype=np.int) # Topic
    W = np.empty((M,N,V),dtype=np.bool) # Document    
    
    # YOUR CODE HERE
    alpha = np.ones(K)*alpha_0
    theta = dirichlet.rvs(alpha, random_state = 42)[0]
    Z = multinomial.rvs(n=1, p=theta, size=(M, N), random_state = 42)
    
    for m in np.arange(M):
        for n in np.arange(N):
            topic = (Z[m,n][Z[m,n] == 1])[0] # select the dimension where it equals one
            W[m,n] = (multinomial.rvs(n=1, p=beta[topic]) != 0)
            
    assert Z.shape == (M,N,K)
    assert W.shape == (M,N,V)
    return W,Z,theta
```
