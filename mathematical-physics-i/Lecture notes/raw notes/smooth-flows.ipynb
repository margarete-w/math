```python
pretty_print_default(True)
var('x_1, x_2, t')
```

```python
## Flow from Sheet 3, Exercis 1

# first component
X = 1 + x_1^2 + x_2^2 + (1 - x_1^2 - x_2^2)*cos(t) - 2*x_2*sin(t)
Phi1(t,x_1, x_2) = 2*x_1 / X
Phi1
```

```python
# second component
Phi2(t, x_1, x_2) = (2*x_2*cos(t) + (1 - x_1^2 - x_2^2)*sin(t))/X
Phi2
```

```python
# computing the corresponding vector field
V1(x_1,x_2) =  diff(Phi1, t)(0, x_1,x_2)
V1
```

```python
V2(x_1, x_2) = diff(Phi2, t)(0, x_1,x_2)
V2
```

```python
# vector filed plot

N = 4
plot1 = plot_vector_field([V1, V2], (x_1,-N,N), (x_2,-N,N))
plot1
```

```python
# solving the ODE numerically to obtain an orbit
sol = desolve_odeint([V1, V2], ics=[1,1], times=srange(0,10,0.01), dvars=[x_1,x_2])
```

```python
# plotting the orbit and the vector field together
plot1 + list_plot(sol)
```
