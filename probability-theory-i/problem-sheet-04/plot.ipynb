```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f_D(x):
    if x < 52.4:
        return 0
    elif x > 52.6:
        return 0
    else:
        return 2.5 * np.pi * np.cos(5 * np.pi*(x - 52.5))
```

```python
dom = np.arange(52.3,52.7,0.001)
y = [f_D(x) for x in dom]
```

```python
plt.plot(dom,y)
```
